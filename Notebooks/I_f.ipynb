# I_f : Hyperpolarization-activated current

```python
import matplotlib.pyplot as plt
import sys
import os
import os
import sys 
import matplotlib.pyplot as plt
from constants import K_o, g_K1,E_K,E_Na,F,R,T,g_f
import numpy as np
```

```python
# Constants
R = 8314.0  # mJ/mol K, Ideal gas constant
T = 295  # K, Absolute temperature
F = 96487.0  # C/mol, Faraday constant
Na_i = 1.073519e+01  # Intracellular sodium concentration in mM
Na_o = 140.0  # Extracellular sodium concentration in mM
K_o = 5.4  # Extracellular potassium concentration in mM
K_i = 1.392751e+02  # Intracellular potassium concentration in mM
g_f = 0.00145  # Maximum conductance for I_f in µS
f_Na = 0.2
f_K = 1 - f_Na

# Equilibrium potentials
E_Na = (R * T / F) * np.log(Na_o / Na_i)  # Sodium equilibrium potential in mV
E_K = (R * T / F) * np.log(K_o / K_i)  # Potassium equilibrium potential in mV

# Steady state value and time constant
def y_inf(V):
    return 1 / (1 + np.exp((V + 138.6) / 10.48))

def tau_y(V):
    return 1 / (0.11885 * np.exp((V + 80.0) / 28.37) + 0.56236 * np.exp((V + 80.0) / -14.19))

# Function to calculate the steady-state I_f
def hypolerisation_current(V, g_f, E_Na, E_K):
    y_ss = y_inf(V)
    I_f = g_f * y_ss * (f_Na * (V - E_Na) + f_K * (V - E_K))
    return I_f

# Example usage
V_range = np.linspace(-100, 50, 300)
I_f_values = [hypolerisation_current(V, g_f, E_Na, E_K) for V in V_range]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(V_range, I_f_values, label='I_f', color='blue')
plt.xlim(-100, 50)
plt.xlabel('Membrane Potential (mV)')
plt.ylabel('Steady State I_f (pA/pF)')
plt.title('Steady State Hyperpolarization-activated Current (I_f) vs Membrane Potential')
plt.legend()
plt.grid(True)
plt.show()
```
